## How to run the model

```python
import tensorflow as tf
model = tf.keras.models.load_model('/content/drive/My Drive/CNN-horse-vs-human/model.keras')
```

```python
from tensorflow.keras.preprocessing import image
import numpy as np

img = image.load_img('/content/drive/My Drive/CNN-horse-vs-human/images/horseIMG.jpeg', target_size=(300, 300))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = img_array / 255.0

prediction = model.predict(img_array)
print("Prediction:", prediction)
```

```python
print("Human" if prediction[0] > 0.5 else "Horse")
```
